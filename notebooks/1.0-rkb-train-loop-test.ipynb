```python
import torch
import numpy as np
from PIL import Image
```

```python
Batch = 4
Channels = 3
Height = 256
Width = 256
img = torch.randn(Batch, Channels, Height, Width)
```

```python
img = img.detach().cpu()
img.shape
```

```python
#img -> (4, 3, 256, 256)
img = (img + 1) * (255 / 2.)
img = img.data.numpy()
img = np.transpose(img, (0, 2, 3, 1))

one_img = img[0]
```

```python
one_img = one_img.astype(np.uint8)
one_img
```

```python
one_img.shape
```

```python
Image.fromarray(one_img)
```
